```python
from PyPDF2 import PdfReader
import os
import json

# path to your dataset folder
dataset_path = r'C:\Dataset\Dataset'

# dictionary to store metadata
metadata_dict = {}

# loop through all author folders
for author_folder in os.listdir(dataset_path):
    author_path = os.path.join(dataset_path, author_folder)

    if os.path.isdir(author_path):
        print(f"\n=== Author: {author_folder} ===")
        author_papers = []  # list to hold all papers for this author

        # loop through all PDFs in the author's folder
        for file in os.listdir(author_path):
            if file.lower().endswith('.pdf'):
                pdf_path = os.path.join(author_path, file)

                try:
                    reader = PdfReader(pdf_path)
                    meta = reader.metadata

                    # convert metadata keys/values to strings (for JSON safety)
                    if meta:
                        clean_meta = {str(k): str(v) for k, v in meta.items()}
                    else:
                        clean_meta = {}

                    author_papers.append({
                        "Paper": file,
                        "Metadata": clean_meta
                    })

                    print(f"  ✅ Extracted: {file}")

                except Exception as e:
                    print(f"  ⚠️ Error reading {file}: {e}")

        metadata_dict[author_folder] = author_papers

# save everything into a JSON file
output_path = os.path.join(dataset_path, "metadata.json")
with open(output_path, "w", encoding="utf-8") as f:
    json.dump(metadata_dict, f, indent=4, ensure_ascii=False)

print(f"\n✅ Metadata extraction complete! Saved to: {output_path}")
```
